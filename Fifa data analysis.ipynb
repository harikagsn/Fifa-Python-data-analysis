### Question 3

Use `groupby` to group the `rating` variable by the player's `position`. Which `position` has the lowest mean `rating`? Remember to chain `.sort_values(ascending=True)` to sort the resulting data on their values and not alphabetically by position name.

```python
q3 = fifa.groupby(['rating', 'position']).mean()
q3.sort_values(by=["rating"], ascending = True)
```

```python
pos = sorted(fifa['rating'])
res3 = fifa.loc[(fifa["rating"] == min(pos)), ["rating", "position"]]
res3
```

### Question 4

Create a Pandas `pivot_table` to count the number of players in each `position`. Use `rating` as the aggregation value again, but this time pass `aggfunc='count'` to report the number of players. Which `position` has the least players? Remember to sort the values in ascending order, and include `by='rating'` to sort on the counted data.

```python
table = pd.pivot_table(fifa, index=['position'], values = ['rating'], aggfunc = 'count')
table.sort_values(by=["rating"], ascending=True)
```

```python
xyz = sorted(table['rating'])
res4 = table.loc[(table["rating"] == min(xyz)), ["rating"]]
res4
```

### Question 7

Group the MLS `base_salary` data by `club` to find the club with the highest average (`mean`) salary. Group the MLS salary data again to find the club with the lowest `min` salary. Is there a difference between the two? Remember to sort the values from smallest to largest.

```python
q7= sal.groupby('club').agg({'base_salary':'mean'})
q7_max = sal.groupby('club')['base_salary'].max().reset_index()
ab = q7_max.sort_values(by=["base_salary"], ascending=True)
ab.max()
```

```python
q7_min = sal.groupby('club')['base_salary'].min().reset_index()
cd = q7_min.sort_values(by=["base_salary"], ascending=True)
cd.min()
```
